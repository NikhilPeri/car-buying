```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('data/toronto_cleaned.csv')
data.columns
```

```python
data['attributes_make_model'] = data['attributes_brand'] + '_' + data['attributes_model']
```

```python
data[data['attributes_vehicle_model_date'] < 2015] \
    .groupby('attributes_vehicle_model_date') \
    .aggregate({'attributes_mileage_from_odometer': 'mean'}) \
    .plot()
```

```python
data.groupby(pd.cut(data['attributes_mileage_from_odometer'], bins=(1000*np.array(range(300))))).aggregate({'price': 'mean'}).plot()
```
